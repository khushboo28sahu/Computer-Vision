# SURF

```
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/drive/My Drive/Mtech_Sem2 /Computer Vision/Colab_FIle/Corner (2).jpg')
img1= cv2.imread('/content/drive/My Drive/Mtech_Sem2 /Computer Vision/Colab_FIle/corner 3.jpg')
img2 = cv2.imread('/content/drive/My Drive/Mtech_Sem2 /Computer Vision/Colab_FIle/corner 4.jpg')

gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
gray_img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

surf = cv2.xfeatures2d.SURF_create()
kp, des = surf.detectAndCompute(gray_img, None)
kp1, des = surf.detectAndCompute(gray_img1, None)
kp2, des = surf.detectAndCompute(gray_img2, None)

kp_img = cv2.drawKeypoints(img, kp, None, color=(0, 255, 0), flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
kp_img1 = cv2.drawKeypoints(img1, kp1, None, color=(0, 255, 0), flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
kp_img2= cv2.drawKeypoints(img2, kp2, None, color=(0, 255, 0), flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)


cv2_imshow(kp_img)
```

```
cv2_imshow(kp_img1)

```

```
cv2_imshow(kp_img2)
cv2.waitKey()
```
