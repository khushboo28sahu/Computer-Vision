# Corner Detection using Shi-Tomasi Corner Detector

```
import numpy as np
import cv2
from matplotlib import pyplot as plt

image = cv2.imread('/content/drive/My Drive/Mtech_Sem2 /Computer Vision/Colab_FIle/Corner (2).jpg')
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

corners = cv2.goodFeaturesToTrack(gray,25,0.01,10)
corners = np.int0(corners)

for i in corners:
    x,y = i.ravel()
    cv2.circle(image,(x,y),3,255,-1)

plt.imshow(image),plt.show()
```

# Corner Detection using Harris Corner Detection

```
import cv2
import numpy as np
from google.colab.patches import cv2_imshow


filename = '/content/drive/My Drive/Mtech_Sem2 /Computer Vision/Colab_FIle/Corner (2).jpg'
img = cv2.imread(filename)
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

gray = np.float32(gray)
dst = cv2.cornerHarris(gray,2,3,0.04)

#result is dilated for marking the corners, not important
dst = cv2.dilate(dst,None)

# Threshold for an optimal value, it may vary depending on the image.
img[dst>0.01*dst.max()]=[0,0,255]
#cv2_imshow(img)
cv2_imshow(dst)
if cv2.waitKey(0) & 0xff == 27:
    cv2.destroyAllWindows()
```

# Corner Detection using FAST Algorithm

```
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('/content/drive/My Drive/Mtech_Sem2 /Computer Vision/Colab_FIle/Corner (2).jpg')
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

fast = cv2.FastFeatureDetector_create()
fast.setNonmaxSuppression(False)

kp = fast.detect(gray_img, None)
kp_img = cv2.drawKeypoints(img, kp, None, color=(0, 255, 0))

cv2_imshow( kp_img)
cv2.waitKey()
```
